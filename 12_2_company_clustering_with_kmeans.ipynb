```
import pandas as pd

def get_fin(stock_code):
  raw = pd.read_html('https://finance.naver.com/item/main.nhn?code={}'.format(stock_code),attrs={'class':'tb_type1 tb_num tb_type1_ifrs'},encoding='cp949')
  fin_df = raw[0]
  fin_df.columns= fin_df.columns.get_level_values(1)
  fin_df = fin_df.set_index('주요재무정보')
  df = fin_df.iloc[:,8]
  df= fin_df.iloc[:,8]
  df.loc['종목코드']=stock_code
  return df
  
```

```
get_fin('005930')
```

```
def get_fins(stock_codes):
  res=[]
  for stock_code in stock_codes:
    print(stock_code)
    df= get_fin(stock_code)
    res.append(df)
  res_df = pd.concat(res, axis=1)
  return res_df
```
